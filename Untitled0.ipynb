```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# 1. CREATE DATA: Let's pretend this is your company's sales data
data = {
    'Product': ['Laptop', 'Monitor', 'Keyboard', 'Laptop', 'Chair', 'Monitor', 'Desk'],
    'Category': ['Tech', 'Tech', 'Tech', 'Tech', 'Office', 'Tech', 'Office'],
    'Sales': [1200, 300, 100, 1200, 450, 300, 800],
    'Profit': [200, 50, 20, 200, 100, 50, 150]
}
df = pd.DataFrame(data)

# 2. SQL STEP: We'll use SQL to find the total profit by category
# We load the data into a temporary SQL database in your browser
conn = sqlite3.connect(':memory:')
df.to_sql('sales_table', conn, index=False)

query = """
SELECT Category, SUM(Profit) as Total_Profit
FROM sales_table
GROUP BY Category
ORDER BY Total_Profit DESC
"""
result = pd.read_sql_query(query, conn)

# 3. SHOW THE PROOF: Print the table and a quick chart
print("--- Business Insight: Profit by Category ---")
print(result)

result.plot(kind='bar', x='Category', y='Total_Profit', color='skyblue')
plt.title('Profitability Analysis for Q1')
plt.ylabel('Profit ($)')
plt.show()
```
